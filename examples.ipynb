```python
import pandas as pd


df = pd.read_csv('3_x_perf_normalization.csv')

df = df[df['performance']<15200]

print(df.shape)
df.head()
```

```python
column_names = df.columns.values
df_grouped = pd.DataFrame(df.groupby(list(column_names[0:7]))['performance'].mean()).reset_index()

df_grouped
```

```python
X_data = df.drop(['filename','failed','performance','perf_norm'],axis=1)
y_data = df['performance']
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

from sklearn import linear_model
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn import neighbors

#training model

cv = 5 # number of folds, i.e. train-test split is 2:1
#model = linear_model.LinearRegression() # you can use any other model here
#model = DecisionTreeRegressor()
#model = RandomForestRegressor()
#model = SVR(gamma='scale', C=1.0, epsilon=0.2)
model = neighbors.KNeighborsRegressor(n_neighbors=5, weights='uniform')

# scores for all folds
scores = cross_val_score(model, X_data, y_data, cv=cv, scoring = 'r2')
print(scores)
print("Accuracy: {} (+/- {})".format(scores.mean(), scores.std() * 2))
```
